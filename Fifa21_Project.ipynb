### Improving the model by modifying the criteria for selecting data (considering individual variables, rather than summations)

```python
df3 = pd.read_csv('./fifa21_train.csv')

def preprocess2(df):
    df = df.drop_duplicates()
    df.columns = df.columns.str.lower().str.replace(' ','_')
    df['ir'] = pd.to_numeric(df['ir'].str.rstrip('★'), errors='coerce')
    desired_columns = ['aggression', 'interceptions', 'positioning',  'vision', 'penalties', 'composure', 'dribbling','curve', 'fk_accuracy', 'long_passing', 'ball_control', 'defending', 'standing_tackle', 'sliding_tackle', 'acceleration', 'sprint_speed', 'agility', 'reactions', 'balance', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes', 'shot_power', 'jumping', 'stamina', 'strength', 'long_shots', 'ir', 'crossing', 'finishing', 'heading_accuracy', 'short_passing', 'volleys', 'ova']
    df = df.loc[:, desired_columns]
    df['interceptions'].fillna(df['interceptions'].mean(), inplace=True)
    df['positioning'].fillna(df['positioning'].mean(), inplace=True)
    df['composure'].fillna(df['composure'].mean(), inplace=True)
    df['vision'].fillna(df['vision'].mean(), inplace=True)
    df['curve'].fillna(df['curve'].mean(), inplace=True)
    df['sliding_tackle'].fillna(df['sliding_tackle'].mean(), inplace=True)
    df['agility'].fillna(df['agility'].mean(), inplace=True)
    df['balance'].fillna(df['balance'].mean(), inplace=True)
    df['jumping'].fillna(df['jumping'].mean(), inplace=True)
    df['volleys'].fillna(df['volleys'].mean(), inplace=True)
    return df

df3 = preprocess2(df3)
df3
```
